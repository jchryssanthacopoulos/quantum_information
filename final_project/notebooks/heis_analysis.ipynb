# Analysis of TEBD on the Heisenberg Model

```python
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm

from tebd.tebd import run_tebd_heis
```

## Effect of $N$

Consider the $\lambda=0$ case.

```python
Ns = range(3, 10)
```

```python
observables = ['energy']
```

```python
energies_N = np.zeros(len(Ns))

for i in tqdm(range(len(Ns))):
    observables_at_midsteps = run_tebd_heis(Ns[i], observables=observables, print_to_stdout=False)
    energies_N[i] = observables_at_midsteps['energy'][-1]
```

```python
plt.plot(Ns, energies_N, 'ro', markersize=3, label="TEBD")
plt.xlabel("Number of sites")
plt.ylabel("$E_0$")
plt.xticks(Ns)
plt.legend()
plt.title("$J_x = J_y = J_z = 1$")
plt.grid()
plt.savefig('../figures/energy_vs_N_lambda_0_heis.png', bbox_inches='tight', dpi=300);
```
